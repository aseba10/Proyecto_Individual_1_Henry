```python
df['genres'] = df['genres'].apply(lambda x: eval(x.strip()) if isinstance(x, str) else x)
df['production_countries'] = df['production_countries'].apply(lambda x: eval(x.strip()) if isinstance(x, str) else x)
df['production_companies'] = df['production_companies'].apply(lambda x: eval(x.strip()) if isinstance(x, str) else x)
```

```python

elementos_unicos_genres = set()

df['genres'].apply(lambda x: elementos_unicos_genres.update(x) if isinstance(x, list) else None)

lista_elementos_unicos_genres = list(elementos_unicos_genres)
lista_elementos_unicos_genres

```

```python
lista_elementos_unicos_genres

contador_genres = {}  # Diccionario para almacenar las cuentas de cada elemento

for elemento in lista_elementos_unicos_genres:
    cuenta = df['genres'].apply(lambda x: x.count(elemento) if isinstance(x, list) else 0).sum()
    contador_genres[elemento] = cuenta

contador_genres
```

```python
elementos_unicos_countries = set()

df['production_countries'].apply(lambda x: elementos_unicos_countries.update(x) if isinstance(x, list) else None)

lista_elementos_unicos_countries = list(elementos_unicos_countries)

contador_countries = {}  # Diccionario para almacenar las cuentas de cada elemento

for elemento in lista_elementos_unicos_countries:
    cuenta = df['production_countries'].apply(lambda x: x.count(elemento) if isinstance(x, list) else 0).sum()
    contador_countries[elemento] = cuenta

contador_countries
```

```python
df_genres = pd.DataFrame.from_dict(contador_genres, orient='index', columns=['Frecuencia'])
top_10_genres = df_genres.nlargest(10, 'Frecuencia')
plt.figure(figsize=(10, 6))
sns.barplot(x='Frecuencia', y=top_10_genres.index, data=top_10_genres, order=top_10_genres.index)
plt.title('Cantidad de Peliculas por Genero')
plt.xlabel('Frecuencia')
plt.ylabel('Genero')

# Ajustar el rango del eje x
plt.xlim(0, top_10_genres['Frecuencia'].max())

plt.show()
```

```python
df_countries = pd.DataFrame.from_dict(contador_countries, orient='index', columns=['Frecuencia'])
top_10_countries = df_countries.nlargest(10, 'Frecuencia')
plt.figure(figsize=(10, 6))
sns.barplot(x='Frecuencia', y=top_10_countries.index, data=top_10_countries, order=top_10_countries.index)
plt.title('Cantidad de Peliculas por Pais')
plt.xlabel('Frecuencia')
plt.ylabel('Pais')

# Ajustar el rango del eje x
plt.xlim(0, top_10_countries['Frecuencia'].max())

plt.show()
```
