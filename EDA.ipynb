```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


titulo = ' '.join(df['title'].astype(str))

# Crear el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(titulo)

# Mostrar la nube de palabras utilizando matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
overview = ' '.join(df['overview'].astype(str))

# Crear el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(overview)

# Mostrar la nube de palabras utilizando matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
tagline = ' '.join(df['tagline'].astype(str).loc[df['tagline'].notna()])

# Crear el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(tagline)

# Mostrar la nube de palabras utilizando matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
